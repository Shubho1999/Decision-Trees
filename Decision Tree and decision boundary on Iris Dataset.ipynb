```python
import numpy as np
import pandas as pd
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris
```

```python
iris.drop('Id',axis=1,inplace=True)
```

```python
iris.drop('SepalWidthCm',axis=1,inplace=True)
```

```python
iris.drop('PetalWidthCm',axis=1,inplace=True)
```

```python
iris.shape
```

```python
iris['Species'] = iris['Species'].map({'Iris-setosa': 0, 'Iris-virginica': 1,'Iris-versicolor':2})
```

```python
iris['Species']=iris['Species'].astype(float)
```

```python
iris
```

```python
X=iris.iloc[:,0:2].values
y=iris.iloc[:,-1].values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
