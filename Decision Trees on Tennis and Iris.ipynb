# Entropy of parent

```python
import math
```

```python
Ed=-(9/14)*math.log2(9/14)-(5/14)*math.log2(5/14)

Ed
```

# Entropy for Sunny in the   Outlook Column

```python
E1=-(2/5)*math.log2(2/5)-(3/5)*math.log2(3/5)

E1
```

# Entropy for Overcast in the outlook column

```python
# As all Overcast has play as Yes hence the entropy is 0  
E2=0
```

# Entropy for Rain in the Outlook column

```python
E3=-(3/5)*math.log2(3/5)-(2/5)*math.log2(2/5)

E3
```

# Weighted Entropy of Children

```python


We1=(5/14)*E1+(4/14)*E2+(5/14)*E3

We1
```



# Information Gain on Outlook Column

```python
IG1=Ed-We1

IG1
```

## Entropy for hot in temperature column

```python
E11=-(2/4)*math.log2(2/4)-(2/4)*math.log2(2/4)

E11
```

## Entropy for mild in teperature column

```python
pd.crosstab(tennis['temp'],tennis['play'])
```

```python
E12=-(4/6)*math.log2(4/6)-(2/6)*math.log2(2/6)

E12
```

## Entropy for cool in temperature column

```python
E13=-(3/4)*math.log2(3/4)-(1/4)*math.log2(1/4)

E13
```

## Weighted Entropy of Temperature column

```python

We2=(4/14)*E11+(6/14)*E12+(4/14)*E13

We2
```


## IG on Temperature

```python
IG2=Ed-We2

IG2
```

# Entropy of  Wind Column

```python
pd.crosstab(tennis['wind'],tennis['play'])
```

```python
E21=-(3/6)*math.log2(3/6)-(3/6)*math.log2(3/6)

E21
```

```python
E22=-(6/8)*math.log2(6/8)-(2/8)*math.log2(2/8)

E22
```

## Weighted Avg on Wind

```python
We3=(6/14)*E21+(8/14)*E22

We3
```

## IG on Wind

```python
IG3=Ed-We3

IG3
```

```python
max(IG1,IG2,IG3)
```
